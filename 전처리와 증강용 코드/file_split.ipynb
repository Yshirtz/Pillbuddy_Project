```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split # ë°ì´í„° ë¶„í• ì— ìœ ìš©í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©

# --- ì„¤ì •ê°’ ---
# ğŸ“ 1. ë°ì´í„°ì…‹ ìµœìƒìœ„ ê²½ë¡œ (ì¦ê°•ì´ ì™„ë£Œëœ í´ë”)
source_base_dir = "D:/AiSchool/1stProject/New Data/ë„¤ì´ë°/Train_augmentation_Crop"

# ğŸ“ 2. ë¶„í• ëœ ë°ì´í„°ë¥¼ ì €ì¥í•  ìµœì¢… ê²½ë¡œ
output_base_dir = 'D:/AiSchool/1stProject/New Data/ë„¤ì´ë°/class'

# ğŸ“ 3. ë¶„í•  ë¹„ìœ¨ ì„¤ì •
TEST_SIZE = 0.2  # í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¹„ìœ¨. ë‚˜ë¨¸ì§€ëŠ” í›ˆë ¨ ë°ì´í„° 
RANDOM_SEED = 42 # ì¬í˜„ì„±ì„ ìœ„í•œ ëœë¤ ì‹œë“œ ì„¤ì •
# ---------------

# ìµœì¢… ì¶œë ¥ ë””ë ‰í† ë¦¬ êµ¬ì¡°:
# output_base_dir/Train/íšŒì‚¬-ì•½ì´ë¦„-íš¨ëŠ¥/...íŒŒì¼ë“¤
# output_base_dir/Test/íšŒì‚¬-ì•½ì´ë¦„-íš¨ëŠ¥/...íŒŒì¼ë“¤

print(f"âœ… ë°ì´í„° ë¶„í•  ì²˜ë¦¬ ì‹œì‘: {source_base_dir}\n")

# ìµœì¢… Train ë° Test í´ë” ê²½ë¡œ ì„¤ì •
train_dir = os.path.join(output_base_dir, 'train')
test_dir = os.path.join(output_base_dir, 'valid')

# Train/Test ìµœìƒìœ„ í´ë” ìƒì„±
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# ì›ë³¸ ìµœìƒìœ„ í´ë” ë‚´ì˜ ëª¨ë“  í•˜ìœ„ í´ë” ì´ë¦„(ì•½ ì´ë¦„) ëª©ë¡ì„ ê°€ì ¸ì˜µë‹ˆë‹¤.
pill_folders = [d for d in os.listdir(source_base_dir) if os.path.isdir(os.path.join(source_base_dir, d))]

if not pill_folders:
    print("âŒ ì˜¤ë¥˜: ì›ë³¸ ìµœìƒìœ„ í´ë” ë‚´ì— ì²˜ë¦¬í•  í•˜ìœ„ í´ë”ê°€ ì—†ìŠµë‹ˆë‹¤.")
    exit()

total_files_copied = 0

for long_folder_name in pill_folders:
    print(f"\n--- ğŸ§ª í´ë” ë¶„í•  ì‹œì‘: {long_folder_name} ---")
    
    # 1. ê²½ë¡œ ì„¤ì •
    source_folder_path = os.path.join(source_base_dir, long_folder_name)
    
    # 2. Train/Test í•˜ìœ„ í´ë” ê²½ë¡œ (ì›ë³¸ í´ë”ëª… ìœ ì§€)
    train_category_dir = os.path.join(train_dir, long_folder_name)
    test_category_dir = os.path.join(test_dir, long_folder_name)
    
    os.makedirs(train_category_dir, exist_ok=True)
    os.makedirs(test_category_dir, exist_ok=True)
    
    # 3. íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    all_files = sorted([f for f in os.listdir(source_folder_path) if os.path.isfile(os.path.join(source_folder_path, f))])

    if not all_files:
        print("   âš ï¸ ê²½ê³ : íŒŒì¼ì´ ì—†ì–´ ìŠ¤í‚µí•©ë‹ˆë‹¤.")
        continue

    # 4. ğŸŒŸ í›ˆë ¨(Train) ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸(Test) ì„¸íŠ¸ë¡œ ëœë¤ ë¶„í• 
    train_files, test_files = train_test_split(
        all_files, 
        test_size=TEST_SIZE, 
        random_state=RANDOM_SEED
    )

    # 5. ë¶„í• ëœ íŒŒì¼ë“¤ì„ ê° í´ë”ë¡œ ë³µì‚¬
    
    # í›ˆë ¨ íŒŒì¼ ë³µì‚¬
    for file_name in train_files:
        src_path = os.path.join(source_folder_path, file_name)
        dst_path = os.path.join(train_category_dir, file_name)
        shutil.copy2(src_path, dst_path)
        total_files_copied += 1
        
    # í…ŒìŠ¤íŠ¸ íŒŒì¼ ë³µì‚¬
    for file_name in test_files:
        src_path = os.path.join(source_folder_path, file_name)
        dst_path = os.path.join(test_category_dir, file_name)
        shutil.copy2(src_path, dst_path)
        total_files_copied += 1
        
    print(f"   âœ… ì´ íŒŒì¼ ìˆ˜: {len(all_files)}ê°œ")
    print(f"   âœ… Train ë³µì‚¬: {len(train_files)}ê°œ (ì•½ {100 - (TEST_SIZE * 100)}%)")
    print(f"   âœ… Test ë³µì‚¬: {len(test_files)}ê°œ (ì•½ {TEST_SIZE * 100}%)")

print("\n---")
print(f"ğŸ‰ ëª¨ë“  í´ë” ë¶„í•  ë° ë³µì‚¬ ì™„ë£Œ. ì´ {total_files_copied}ê°œì˜ íŒŒì¼ì´ Train/Testë¡œ ë¶„í•  ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
