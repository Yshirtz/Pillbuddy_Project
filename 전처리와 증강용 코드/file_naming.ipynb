```python
import os
import shutil
import re

# --- ì„¤ì •ê°’ ---
# ğŸ“ ì›ë³¸ ë°ì´í„°ì…‹ì˜ ìµœìƒìœ„ í´ë” ê²½ë¡œ
source_base_dir = 'D:/AiSchool/1stProject/New Data/ë„¤ì´ë°/Train_naming'
# ğŸ“ ê²°ê³¼ íŒŒì¼ì„ ì €ì¥í•  ìµœìƒìœ„ í´ë” ê²½ë¡œ
new_base_dir = 'D:/AiSchool/1stProject/New Data/ë„¤ì´ë°/Train_augmentation'

new_suffix = 'original' # íŒŒì¼ëª…ì— ë¶™ì„ ì ‘ë¯¸ì‚¬
# ---------------

print(f"âœ… ë°ì´í„° ì²˜ë¦¬ ì‹œì‘: {source_base_dir}\n")

# ì›ë³¸ ìµœìƒìœ„ í´ë” ë‚´ì˜ ëª¨ë“  í•˜ìœ„ í´ë” ì´ë¦„(ì•½ ì´ë¦„) ëª©ë¡ì„ ê°€ì ¸ì˜µë‹ˆë‹¤.
pill_folders = [d for d in os.listdir(source_base_dir) if os.path.isdir(os.path.join(source_base_dir, d))]

if not pill_folders:
    print("âŒ ì˜¤ë¥˜: ì›ë³¸ ìµœìƒìœ„ í´ë” ë‚´ì— ì²˜ë¦¬í•  í•˜ìœ„ í´ë”ê°€ ì—†ìŠµë‹ˆë‹¤.")
    exit()

total_files_processed = 0

for long_folder_name in pill_folders:
    print(f"\n--- ğŸ§ª í´ë” ì²˜ë¦¬ ì‹œì‘: {long_folder_name} ---")
    
    # 1. ê²½ë¡œ ì„¤ì •
    source_folder_path = os.path.join(source_base_dir, long_folder_name)
    new_folder_path = os.path.join(new_base_dir, long_folder_name)
    
    # 2. ğŸŒŸ í´ë” ì´ë¦„ì—ì„œ í•µì‹¬ ì•½ ì´ë¦„ë§Œ ì¶”ì¶œ (ê°€ìš´ë° ë¶€ë¶„)
    parts = long_folder_name.split('_')
    
    if len(parts) >= 2:
        # í•˜ì´í”ˆìœ¼ë¡œ ë¶„ë¦¬ëœ ë¶€ë¶„ ì¤‘ ë‘ ë²ˆì§¸ ìš”ì†Œ(ì¸ë±ìŠ¤ 1)ë¥¼ í•µì‹¬ ì•½ ì´ë¦„ìœ¼ë¡œ ì‚¬ìš©
        core_name_prefix = parts[1].strip()
    else:
        # ê·œì¹™ì— ë§ì§€ ì•Šì„ ê²½ìš° í´ë” ì´ë¦„ì„ ê·¸ëŒ€ë¡œ ì‚¬ìš©
        core_name_prefix = long_folder_name.strip()
        
    if not core_name_prefix:
        core_name_prefix = "NewFiles"
        
    # ëŒ€ìƒ í´ë” ìƒì„± (ê° ì•½ ì´ë¦„ í´ë”ë³„ë¡œ)
    os.makedirs(new_folder_path, exist_ok=True)
    print(f"   âœ… íŒŒì¼ëª… ì ‘ë‘ì‚¬: {core_name_prefix}")
    print(f"   âœ… ì €ì¥ë  í´ë”: {new_folder_path}")

    # íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
    file_names = sorted([f for f in os.listdir(source_folder_path) if os.path.isfile(os.path.join(source_folder_path, f))])

    i = 1 # ë„˜ë²„ë§ ì‹œì‘ ê°’ (ê° í´ë”ë§ˆë‹¤ ì´ˆê¸°í™”)

    for old_name in file_names:
        # 3. íŒŒì¼ëª…ê³¼ í™•ì¥ì ë¶„ë¦¬
        match = re.search(r'(\.[^.]+)$', old_name)
        ext_part = match.group(1) if match else '' 
        
        # 4. ìƒˆë¡œìš´ íŒŒì¼ëª… êµ¬ì„±
        new_number = str(i).zfill(3) # 3ìë¦¬ ë„˜ë²„ë§
        
        # ìµœì¢… ìƒˆë¡œìš´ íŒŒì¼ëª…: (í•µì‹¬ ì•½ ì´ë¦„)_(ë„˜ë²„ë§)_original.(í™•ì¥ì)
        new_name = f"{core_name_prefix}_{new_number}_{new_suffix}{ext_part}"

        # ì›ë³¸ íŒŒì¼ ê²½ë¡œ
        src_file = os.path.join(source_folder_path, old_name)
        # ìƒˆë¡œìš´ íŒŒì¼(ë³µì‚¬ë³¸)ì´ ì €ì¥ë  ê²½ë¡œ
        dst_file = os.path.join(new_folder_path, new_name)

        # 5. íŒŒì¼ ë³µì‚¬ ë° ì´ë¦„ ë³€ê²½ ì‹¤í–‰
        try:
            # í•œê¸€ ê²½ë¡œë¥¼ ì•ˆì „í•˜ê²Œ ì²˜ë¦¬í•˜ëŠ” shutil.copy2 ì‚¬ìš©
            shutil.copy2(src_file, dst_file)
            # print(f"   {old_name} -> {new_name}") # íŒŒì¼ì´ ë„ˆë¬´ ë§ì„ ê²½ìš° ì£¼ì„ ì²˜ë¦¬ ê°€ëŠ¥
            i += 1
            total_files_processed += 1
        except Exception as e:
            print(f"âŒ ì˜¤ë¥˜ ë°œìƒ ({long_folder_name}/{old_name}): {e}")

    print(f"   âœ… ì²˜ë¦¬ëœ íŒŒì¼ ìˆ˜: {i - 1}ê°œ")

print("\n---")
print(f"ğŸ‰ ëª¨ë“  í´ë” ì²˜ë¦¬ ì™„ë£Œ. ì´ {total_files_processed}ê°œì˜ íŒŒì¼ì´ ë³µì‚¬ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
